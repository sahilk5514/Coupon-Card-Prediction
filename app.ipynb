```python
import streamlit as st
import requests

# Set your actual API endpoint here
API_URL = 'YOUR_API_ENDPOINT_HERE'

# Page config
st.set_page_config(page_title="Coupon Redemption Predictor", layout="centered")

# Background styling using custom CSS
st.markdown("""
    <style>
        body {
            background-color: #000000;
        }
        .main {
            background: url('background.png') no-repeat center center fixed;
            background-size: cover;
        }
        .stApp {
            background-color: #000000;
        }
        .stTextInput>div>div>input {
            background-color: #ffffff;
            color: #000000;
        }
    </style>
""", unsafe_allow_html=True)

# Main Container
with st.container():
    st.markdown(
        "<h1 style='text-align: center; color: white; background-color: #b42324; padding: 15px; border-radius: 10px;'>Coupon Redemption Predictor</h1>",
        unsafe_allow_html=True
    )

    with st.form("prediction_form"):
        age = st.number_input("Customer Age", min_value=0, max_value=120, step=1)
        past_redemption = st.slider("Past Redemption Frequency (0-10)", 0, 10)
        discount = st.slider("Coupon Discount (%)", 0, 100)
        validity = st.number_input("Coupon Validity (Days)", min_value=0, step=1)
        offer_type = st.selectbox("Offer Type", ["", "Discount", "Cashback"])

        submitted = st.form_submit_button("Predict Redemption")

    if submitted:
        if offer_type == "":
            st.warning("Please select an offer type.")
        else:
            formData = {
                "age": age,
                "past_redemption": past_redemption,
                "discount": discount,
                "validity": validity,
                "offer_type": offer_type
            }

            try:
                response = requests.post(API_URL, json=formData)
                if response.status_code == 200:
                    data = response.json()
                    prediction = data.get("prediction", [None])[0]
                    confidence = data.get("confidence", [None])[0]

                    if prediction is not None:
                        if prediction == 1:
                            st.success(f"üéâ Predicted Redemption: Likely to Redeem\nConfidence: {confidence * 100:.2f}%")
                        else:
                            st.error(f"‚ùå Predicted Redemption: Unlikely to Redeem\nConfidence: {confidence * 100:.2f}%")
                    else:
                        st.warning("Prediction not available.")
                else:
                    st.error(f"Server error! Status code: {response.status_code}")
            except Exception as e:
                st.error(f"An error occurred: {e}")
```
